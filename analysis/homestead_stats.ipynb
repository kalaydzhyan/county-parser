```python
import pandas as pd
import numpy as np
import collections
```

```python
def owner_percentage(df):
    counter        = collections.Counter(df.absentee)
    owner_occupied = counter[False]
    total_homes    = len(df)
    return round(100*owner_occupied/total_homes)
```

```python
#df = pd.read_csv('/Users/tigrank/Downloads/output_taylor (4).csv')
df = pd.read_csv('../output/output_taylor/output_taylor.csv')
```

```python
df_filtered = df[(df.improvement_value>10000) & (df.zoning=='Residential') & (np.isin(df.school.values, ['ABILENE', 'WYLIE']))]
owner_percentage(df_filtered)
```

```python
df_green_acres = df_filtered[df_filtered.legal_description.str.startswith("GREEN ACRES")]
owner_percentage(df_green_acres)
```

```python
df_park_plaza = df_filtered[df_filtered.legal_description.str.startswith("PARK PLAZA")]
owner_percentage(df_park_plaza)
```
